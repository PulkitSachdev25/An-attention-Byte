```python
from pydub import AudioSegment
from pydub.utils import make_chunks

audio = AudioSegment.from_wav("myaudio.wav")

CLIP_MIN_PEAKS = 5
CLIP_MAX_PEAKS = 6
THRESHOLD_START = -15
THRESHOLD_MIN = -40
THRESHOLD_STEP = 1
CHUNK_MS = 1000  # 1 second chunks now
MERGE_GAP = 3000  # merge peaks within 3 seconds

threshold_db = THRESHOLD_START
success = False

while threshold_db >= THRESHOLD_MIN:
    chunks = make_chunks(audio, CHUNK_MS)
    peaks = []

    # Find peaks above threshold
    for i, chunk in enumerate(chunks):
        if chunk.dBFS >= threshold_db:
            peaks.append(i*CHUNK_MS)

    # Merge nearby peaks into longer segments
    merged = []
    if peaks:
        start = peaks[0]
        end = peaks[0]
        for t in peaks[1:]:
            if t - end <= MERGE_GAP:
                end = t
            else:
                merged.append([start/1000, (end + CHUNK_MS)/1000])  # extend to end of last chunk
                start = end = t
        merged.append([start/1000, (end + CHUNK_MS)/1000])
    else:
        merged = []

    print(f"Threshold: {threshold_db} dB | Merged peaks: {len(merged)}")

    if CLIP_MIN_PEAKS <= len(merged) <= CLIP_MAX_PEAKS:
        success = True
        break
    elif len(merged) < CLIP_MIN_PEAKS:
        threshold_db -= THRESHOLD_STEP
    else:
        threshold_db += THRESHOLD_STEP

if not success:
    print("Video too quiet or peaks not detected properly!")
else:
    print("Final merged peaks (seconds):", merged)
```

```python
transcript = []
for segment in segments:
    transcript.append({
        "start": segment.start,
        "end": segment.end,
        "text": segment.text.strip()
    })
```

```python
import re

# transcript = your Whisper output: list of dicts with start, end, text
# Example:
# transcript = [{'start': 10.64, 'end': 15.76, 'text': 'The crowd holds its breath ...'}, ...]

full_text = ""
word_to_timestamp = []  # store start and end time of each word

for segment in transcript:
    words = segment["text"].split()  # split segment into words
    num_words = len(words)
    segment_duration = segment["end"] - segment["start"]

    for i, word in enumerate(words):
        # estimate timestamp of each word linearly within segment
        word_start = segment["start"] + (i / num_words) * segment_duration
        word_end   = segment["start"] + ((i+1) / num_words) * segment_duration
        word_to_timestamp.append((word, word_start, word_end))
        full_text += word + " "

full_text = full_text.strip()
```

```python
def expand_peak_to_sentence(peak_time, word_to_timestamp, full_text):
    # 1. Find the word closest to peak
    closest_idx = min(range(len(word_to_timestamp)), key=lambda i: abs(word_to_timestamp[i][1] - peak_time))

    # 2. Find sentence start
    sentence_start_idx = closest_idx
    while sentence_start_idx > 0:
        if re.search(r'[.!?]$', word_to_timestamp[sentence_start_idx-1][0]):
            break
        sentence_start_idx -= 1

    # 3. Find sentence end
    sentence_end_idx = closest_idx
    while sentence_end_idx < len(word_to_timestamp)-1:
        if re.search(r'[.!?]$', word_to_timestamp[sentence_end_idx][0]):
            break
        sentence_end_idx += 1

    # 4. Get start/end times
    start_time = word_to_timestamp[sentence_start_idx][1]
    end_time   = word_to_timestamp[sentence_end_idx][2]

    # 5. Get text
    sentence_text = " ".join([w[0] for w in word_to_timestamp[sentence_start_idx:sentence_end_idx+1]])

    return start_time, end_time, sentence_text
```

```python
# Example: merged = list of peak timestamps in seconds
final_clips = []

for peak_start, peak_end in merged:
    peak_time = (peak_start + peak_end)/2  # middle of peak
    start, end, text = expand_peak_to_sentence(peak_time, word_to_timestamp, full_text)

    final_clips.append({
        "start": start,
        "end": end,
        "text": text
    })

# Check output
for clip in final_clips:
    print(clip)
```

```python
clean_clips = []
for clip in final_clips:
    if not clean_clips:
        clean_clips.append(clip)
    else:
        prev = clean_clips[-1]
        # skip clip if it starts within previous clip
        if clip["start"] <= prev["end"]:
            continue
        clean_clips.append(clip)

# Step 4: output
print("Final clips ready for video extraction:")
for i, clip in enumerate(clean_clips, 1):
    print(f"Clip {i}: {clip['start']:.2f}s - {clip['end']:.2f}s | {clip['text']}")
```

```python

import cv2
import numpy as np
from moviepy.editor import VideoFileClip, concatenate_videoclips, vfx



video_path = "test.mp4"
video = VideoFileClip(video_path)
clips = []

FADE_IN = 0.5
FADE_OUT = 0.5

def add_subtitle_opencv(clip, text):
    """Draw text on each frame using OpenCV"""
    def annotate_frame(frame):
        # Convert float64 frame to uint8
        frame_uint8 = np.clip(frame, 0, 255).astype(np.uint8)
# Convert to BGR for OpenCV
        frame_bgr = cv2.cvtColor(frame_uint8, cv2.COLOR_RGB2BGR)

        # Text parameters
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 0.8
        font_thickness = 2
        text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
        text_x = int((frame.shape[1] - text_size[0]) / 2)
        text_y = frame.shape[0] - 50  # 50 px from bottom
        # Draw background rectangle
        cv2.rectangle(frame_bgr,
                      (text_x - 10, text_y - text_size[1] - 10),
                      (text_x + text_size[0] + 10, text_y + 10),
                      (0, 0, 0), -1)
        # Draw text
        cv2.putText(frame_bgr, text, (text_x, text_y), font, font_scale, (255, 255, 255), font_thickness, cv2.LINE_AA)
        # Convert back to RGB
        frame_rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
        return frame_rgb

    return clip.fl_image(annotate_frame)

for clip_info in clean_clips:
    subclip = video.subclip(clip_info["start"], clip_info["end"]).fadein(FADE_IN).fadeout(FADE_OUT)
    subclip = add_subtitle_opencv(subclip, clip_info["text"])
    clips.append(subclip)

final_video = concatenate_videoclips(clips, method="compose")
final_video.write_videofile("final_mvp_with_subtitles.mp4", fps=24)
```
